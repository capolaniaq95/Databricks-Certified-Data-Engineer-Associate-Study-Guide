```python
%sql

DROP TABLE IF EXISTS hive_metastore.default.product_info;

CREATE TABLE hive_metastore.default.product_info(
  product_id BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
  product_name STRING,
  category STRING,
  price DOUBLE,
  quantity INT
) USING DELTA;
```

```python
%sql
CREATE TABLE hive_metastore.default.user_info(
  user_id BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
  user_name STRING,
  last_name STRING,
  age INT
) USING DELTA;
```
